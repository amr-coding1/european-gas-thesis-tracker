## Background: How European Gas Works

Europe heats its homes, runs its power plants, and fuels its industry with natural gas. Unlike the US (which produces most of its own gas), Europe **imports** the majority of its supply — historically from Russia via pipeline, and increasingly from global LNG (Liquefied Natural Gas) shipped by tanker from Qatar, the US, and elsewhere.

**TTF (Title Transfer Facility)** is the European gas benchmark, priced in the Netherlands. It is to European gas what Brent is to oil — the reference price that all contracts, utilities, and industrial buyers anchor to. When we say "European gas prices," we mean TTF.

**Gas storage** is critical because demand is seasonal — Europe burns far more gas in winter (heating) than summer. To survive each winter, Europe fills underground storage caverns during the summer "injection season" (April-October), then draws them down during winter "withdrawal season" (November-March). The EU passed a law after the 2022 energy crisis requiring member states to fill storage to at least **90%** capacity by November each year (since amended — see below).

## The Event: What Happened at Ras Laffan

On 18-19 March 2026, Iranian missile strikes hit Qatar's **Ras Laffan Industrial City** — the single largest LNG production complex on Earth, responsible for roughly **20% of global LNG supply**.

The strikes destroyed **Train 4** and **Train 6** — two of the facility's liquefaction units. A "train" is a self-contained processing line that cools natural gas to -162°C, converting it to liquid for shipping. These are among the most complex pieces of industrial engineering ever built.

**The damage:**
- **12.8 million tonnes per annum (mtpa)** of LNG capacity destroyed — approximately 17% of Qatar's total output
- **$20 billion per year** in lost revenue (QatarEnergy's own estimate)
- **3-5 years** to repair — liquefaction trains require specialised equipment (cryogenic heat exchangers, gas turbines) with 18-24 month lead times. Money cannot buy time on engineering constraints.
- **Force majeure** declared on long-term supply contracts to **Italy, Belgium, South Korea, and China**
- Shell (joint venture partner) also declared force majeure

**Why this matters more than a typical disruption:** This is **physical destruction of production infrastructure**, not a sanctions regime or shipping blockade that can be reversed by political decision. Even if every geopolitical tension resolved tomorrow, those 12.8 mtpa of LNG physically cannot be produced until the trains are rebuilt. The molecules do not exist.

Additionally, the **Pearl GTL facility** (operated by Shell) at Ras Laffan lost one of its two trains, expected offline for at least one year.

The strait is a narrow waterway through which approximately **20% of global oil** and a significant portion of global LNG transits.

On 5 March, Iran shifted to a **selective blockade** — closed to US, Israeli, and Western-allied vessels, but allowing ships from India, China, Pakistan, and Malaysia to pass. Iran is developing a formal vetting system for transit.

On 19 March, the **US began an aerial campaign** to reopen the strait.

**Impact on the thesis:** Even Qatar's *remaining* operational LNG trains cannot freely export while Hormuz is contested. This means the supply disruption is not limited to the destroyed trains — it extends to all Gulf LNG shipments to Western buyers.

## The Core Thesis

European gas storage is critically low, and Europe is about to enter "injection season" — the period when it must refill its underground storage to survive next winter. The maths required to refill storage does not work without Qatari LNG, and there is no substitute supply source large enough to fill the gap. Therefore, **TTF (the European gas price) must rise significantly** to:

1. **Attract marginal LNG cargoes from Asia** — US LNG tankers go to whoever pays the most. If Europe needs them more than Asia, TTF must trade at a premium to JKM (the Asian LNG benchmark).
2. **Ration European demand** — at sufficiently high prices, energy-intensive industries (fertilisers, chemicals, glass, steel) shut down, reducing gas consumption and making the refill achievable with less supply.

This is the commodity market's fundamental balancing mechanism: when supply is constrained and inventories are low, **price must rise until supply and demand rebalance.**

## Why the Thesis Is Bullish (The Case FOR Higher TTF)

### 1. The refill maths does not work
EU storage is at approximately **28.5%**. The EU mandate requires at least 80% (recommended) to 90% by December 1. Even at the loosest possible interpretation (70% floor), Europe needs to inject at **2-3x its current daily rate, every single day, for 7+ months**. This has never been achieved in any prior injection season, and that was *with* Qatari LNG flowing freely.

### 2. Netherlands storage is nearly empty
The Netherlands — where TTF is physically priced — is at approximately **7% storage**. This translates to roughly **34 days of gas remaining** at current withdrawal rates. The country where Europe's benchmark gas price is set is running on fumes.

### 3. There is no swing supplier
- **Norway:** Pipeline gas is running at maximum. North Sea fields are mature and declining.
- **US LNG:** The largest source of flexible supply, but terminals are already operating near capacity. New capacity (Golden Pass, Plaquemines) is ramping but won't be fully online for summer 2026. And cargoes go to the highest bidder — Asia is competing for the same ships.
- **Qatar (remaining):** Hormuz blockade constrains exports even from undamaged trains.
- **Russia:** The EU formally banned Russian gas imports in January 2026, with full phase-out by 2027. Nord Stream pipelines are physically destroyed. Ukraine transit expired. Only TurkStream (~15 bcm/yr) still flows.
- **Algeria/Libya:** Small pipeline volumes to Southern Europe. Unreliable and declining.
- **Domestic EU production:** Groningen (Netherlands) permanently shut. UK North Sea declining 5-8% per year.

### 4. The EU banned Russian gas — permanently
In January 2026, the EU Council adopted a regulation prohibiting Russian LNG and pipeline gas imports, with full phase-out by end of 2027. This was previously the **#1 bear risk** to this thesis — the possibility that sanctions could be eased if prices got too painful. That option is now closed by law. Russia has responded by threatening to cut off remaining flows preemptively.

### 5. Asia is competing for the same LNG cargoes
Qatar's force majeure didn't just hit European buyers — it also hit **South Korea and China**, the world's 2nd and 3rd largest LNG importers. Both are now scrambling for replacement cargoes from the same pool of US and Australian LNG that Europe needs. This is a global tightening, not just a European one.

### 6. The price confirms the thesis
TTF has already moved from ~€30-35/MWh (pre-strikes) to ~€59/MWh. The market is beginning to price the supply shock, but the full implications of the refill maths haven't been tested yet — that comes when injection season data starts printing in April.

### 7. Theory of storage supports the trade
When inventories are critically low, the **convenience yield** (the value of having physical gas available right now) spikes. This pushes spot prices above futures prices (backwardation), incentivising immediate delivery and discouraging storage — which paradoxically makes the refilling problem worse. It's a self-reinforcing cycle that persists until the physical market rebalances.

## Why the Thesis Could Be Wrong (The Case AGAINST)

### 1. Ceasefire and Hormuz reopening
The US aerial campaign could succeed in reopening the Strait, or diplomatic channels could produce a ceasefire. If Hormuz reopens fully, Qatar's undamaged trains can resume exports and the supply picture improves significantly — though the 12.8 mtpa of destroyed capacity remains offline regardless.

### 2. EU lowers the storage mandate further
The EU has already shown flexibility — recommending 80% instead of 90%, with a floor as low as 70% and a deadline pushed to December 1. If they drop the target further (say, to 60%), the urgency and panic premium in TTF collapses. Politicians may accept lower storage if the alternative is factory closures and voter backlash from high energy bills.

### 3. Demand destruction
High gas prices are self-correcting. At €80-100/MWh, European fertiliser plants, chemical facilities, and heavy industry shut down. This reduces gas demand, making the storage refill easier. The question is not *whether* demand destruction happens, but *at what price level* — and whether that level is above or below the current price. This risk is also partially the thesis itself: TTF needs to rise to the level that triggers demand destruction.

### 4. US LNG capacity additions
New US export terminals (Golden Pass, Plaquemines expansion) are ramping through 2026. If they reach full production faster than expected, there are more cargoes available globally. However, this new capacity was largely pre-contracted (much of it to Asia), and even at full output it doesn't fully replace 12.8 mtpa of Qatari losses on top of the existing Russian gap.

### 5. Mild summer weather
A cool European summer significantly reduces gas demand — less air conditioning load means less gas burn in power plants. In a mild summer, the refill maths becomes more achievable. Weather is the highest-variance unknown and cannot be modelled. A 2-standard-deviation cool summer combined with maximum US LNG flows could potentially close the gap.

### 6. Russian LNG shadow flows
Despite the import ban, Russian LNG from Yamal and Arctic LNG-2 is already reaching Europe through ship-to-ship transfers, dark fleet tankers, and rebranding. If these shadow flows are significantly larger than official statistics suggest, the actual supply gap may be smaller than it appears. Enforcement could also lag behind legislation.

## 1. Live European Gas Storage

```python
# ── Pull live storage data from AGSI API ──

def fetch_storage(country_code, date=None):
    params = dict(COUNTRIES[country_code]['param'])
    if date:
        params['date'] = date
    try:
        r = requests.get(AGSI_URL, headers=HEADERS, params=params, timeout=15)
        r.raise_for_status()
        data = r.json()
        if isinstance(data, list) and data:
            return data[0]
        if isinstance(data, dict) and 'data' in data:
            entries = data['data']
            if isinstance(entries, list) and entries:
                return entries[0]
        return data
    except Exception as e:
        print(f'  Error fetching {country_code}: {e}')
        return None

def safe_float(v):
    if v is None or v == '' or v == '-': return None
    try: return float(v)
    except: return None

# Fetch current data
import time
storage = {}
for i, code in enumerate(COUNTRIES):
    if i > 0: time.sleep(0.5)
    d = fetch_storage(code)
    if d:
        storage[code] = {
            'name': COUNTRIES[code]['name'],
            'full_pct': safe_float(d.get('full')),
            'gas_twh': safe_float(d.get('gasInStorage')),
            'capacity_twh': safe_float(d.get('workingGasVolume')),
            'injection': safe_float(d.get('injection')),
            'withdrawal': safe_float(d.get('withdrawal')),
            'net_withdrawal': safe_float(d.get('netWithdrawal')),
            'gas_day': d.get('gasDayStart', '?'),
        }

# Fetch 7-day-ago for trajectory
week_ago = (datetime.now() - timedelta(days=7)).strftime('%Y-%m-%d')
time.sleep(0.5)
prev_eu = fetch_storage('EU', week_ago)
time.sleep(0.5)
prev_nl = fetch_storage('NL', week_ago)

prev_eu_pct = safe_float(prev_eu.get('full')) if prev_eu else None
prev_nl_pct = safe_float(prev_nl.get('full')) if prev_nl else None

# Display
gas_day = storage.get('EU', {}).get('gas_day', '?')
display(Markdown(f'### Storage Snapshot — Gas Day: {gas_day}'))

rows = []
for code, d in storage.items():
    pct = d['full_pct']
    threshold = COUNTRIES[code]['threshold']
    flag = '⚠️ CRITICAL' if pct is not None and pct < threshold else ''
    nw = d['net_withdrawal']
    rows.append({
        'Country': d['name'],
        'Storage %': f"{pct:.2f}%" if pct else 'N/A',
        'Gas (TWh)': f"{d['gas_twh']:.1f}" if d['gas_twh'] else 'N/A',
        'Capacity (TWh)': f"{d['capacity_twh']:.1f}" if d['capacity_twh'] else 'N/A',
        'Net W/D (GWh/d)': f"{nw:+,.0f}" if nw else 'N/A',
        'Status': flag,
    })

df_storage = pd.DataFrame(rows)
display(df_storage.style.set_properties(**{'text-align': 'right'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'center'), ('font-weight', 'bold')]}]
).hide(axis='index'))

# Weekly change
eu_pct = storage.get('EU', {}).get('full_pct')
nl_pct = storage.get('NL', {}).get('full_pct')

changes = []
if eu_pct and prev_eu_pct:
    delta = eu_pct - prev_eu_pct
    changes.append(f"- **EU:** {prev_eu_pct:.2f}% → {eu_pct:.2f}% ({delta:+.2f}pp {'▼' if delta < 0 else '▲'})")
if nl_pct and prev_nl_pct:
    delta = nl_pct - prev_nl_pct
    changes.append(f"- **Netherlands:** {prev_nl_pct:.2f}% → {nl_pct:.2f}% ({delta:+.2f}pp {'▼' if delta < 0 else '▲'})")

if changes:
    display(Markdown('### Weekly Change (7 days)\n' + '\n'.join(changes)))
```

## 2. The Refill Maths

```python
# ── Refill calculations for different mandate targets ──

eu = storage.get('EU', {})
eu_pct = eu.get('full_pct') or 0
eu_cap = eu.get('capacity_twh') or 1142
current_injection = eu.get('injection') or 0

days_left = max((MANDATE_DEADLINE - datetime.now()).days, 1)

targets = [MANDATE_TARGET, MANDATE_RECOMMENDED, 75, MANDATE_FLOOR]
rows = []
for target in targets:
    gap_pct = target - eu_pct
    if gap_pct <= 0:
        continue
    twh_needed = (gap_pct / 100) * eu_cap
    daily_needed = (twh_needed * 1000) / days_left
    ratio = daily_needed / current_injection if current_injection > 0 else float('inf')
    rows.append({
        'Target': f"{target}%",
        'TWh to Inject': f"{twh_needed:.0f}",
        'Days Left': days_left,
        'Required (GWh/d)': f"{daily_needed:,.0f}",
        'Current (GWh/d)': f"{current_injection:,.0f}" if current_injection > 0 else 'N/A',
        'Gap Ratio': f"{ratio:.1f}x" if ratio != float('inf') else 'N/A',
    })

display(Markdown(f'### EU Refill Maths — {days_left} days to December 1 deadline'))
display(pd.DataFrame(rows).style.hide(axis='index'))

# Netherlands days-to-empty
nl = storage.get('NL', {})
nl_gas = nl.get('gas_twh')
nl_nw = nl.get('net_withdrawal')
if nl_gas and nl_nw and nl_nw > 0:
    dte = (nl_gas * 1000) / nl_nw
    display(Markdown(f'### Netherlands: **{dte:.0f} days** of storage remaining at current withdrawal rate'))
```

## 3. 30-Day EU Storage Trend

```python
# ── Pull 30-day history for EU and NL ──

def fetch_history(code, days=30):
    records = []
    for i in range(days):
        date = (datetime.now() - timedelta(days=i)).strftime('%Y-%m-%d')
        d = fetch_storage(code, date)
        if d:
            pct = safe_float(d.get('full'))
            if pct is not None:
                records.append({'date': pd.Timestamp(date), 'full_pct': pct})
        time.sleep(0.3)
    if not records:
        return pd.DataFrame(columns=['date', 'full_pct'])
    return pd.DataFrame(records).sort_values('date').reset_index(drop=True)

print('Fetching 30-day EU history...')
eu_hist = fetch_history('EU', 30)
print('Fetching 30-day NL history...')
nl_hist = fetch_history('NL', 30)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

# EU chart
if not eu_hist.empty:
    ax1.plot(eu_hist['date'], eu_hist['full_pct'], 'b-', linewidth=2, label='EU Storage %')
    ax1.axhline(y=90, color='green', linestyle='--', alpha=0.5, label='90% mandate')
    ax1.axhline(y=80, color='orange', linestyle='--', alpha=0.5, label='80% (recommended)')
    ax1.axhline(y=70, color='red', linestyle='--', alpha=0.5, label='70% (floor)')
    ax1.fill_between(eu_hist['date'], eu_hist['full_pct'], alpha=0.15, color='blue')
    ax1.set_ylabel('Storage %')
    ax1.set_title('EU Aggregate Storage (30 days)', fontweight='bold')
    ax1.legend(loc='upper right', fontsize=9)
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))
    ax1.set_ylim(0, 100)

# NL chart
if not nl_hist.empty:
    ax2.plot(nl_hist['date'], nl_hist['full_pct'], 'r-', linewidth=2, label='Netherlands Storage %')
    ax2.fill_between(nl_hist['date'], nl_hist['full_pct'], alpha=0.15, color='red')
    ax2.set_ylabel('Storage %')
    ax2.set_title('Netherlands Storage — Where TTF is Priced (30 days)', fontweight='bold')
    ax2.legend(loc='upper right', fontsize=9)
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))
    ax2.set_ylim(0, max(nl_hist['full_pct'].max() * 1.5, 20))

plt.tight_layout()
plt.show()
```

### The Henry Hub Hedge -- How It Works

The short HH is not a perfect hedge -- it's a **decorrelation play**:

- **US gas is structurally oversupplied.** Production at record highs, export terminals maxed out. HH can't rally even if global gas rips.
- **The TTF-HH spread (~EUR 12/MMBtu equivalent) is the trade itself.** Cheniere and VG profit from this spread. Shorting HH is betting the US side stays low.
- **In a ceasefire scenario:** TTF drops, Brent drops, equities drop. But HH was already at $2.94 -- it has very little downside. The short makes a small gain while everything else loses.
- **Honest assessment:** The hedge covers maybe 5% of portfolio risk in the worst scenario. It's better than nothing, and it shows recruiter-grade portfolio thinking. A real hedge would be put options on TTF or Brent.

```python
# ── Live news feed — Google News RSS ──
import time
import urllib.parse
from datetime import datetime, timedelta, timezone

try:
    import feedparser
    HAS_FEEDPARSER = True
except ImportError:
    HAS_FEEDPARSER = False
    print("Install feedparser for live news: pip install feedparser")

NEWS_QUERIES = {
    "Gas / LNG": [
        "TTF gas price Europe",
        "Ras Laffan LNG Qatar repair",
        "European gas storage injection 2026",
        "LNG cargo spot market Asia Europe",
        "Strait of Hormuz shipping LNG",
    ],
    "Geopolitical": [
        "Iran war ceasefire negotiations 2026",
        "Trump Iran deadline energy",
        "EU Russian gas ban sanctions",
    ],
    "Equities": [
        "Cheniere Energy LNG stock",
        "Venture Global LNG Plaquemines",
        "Equinor gas production Norway",
    ],
    "Oil": [
        "Brent crude oil price Hormuz",
    ],
}

if HAS_FEEDPARSER:
    all_news = []
    for category, queries in NEWS_QUERIES.items():
        for query in queries:
            try:
                url = f"https://news.google.com/rss/search?q={urllib.parse.quote(query)}&hl=en&gl=US&ceid=US:en"
                feed = feedparser.parse(url)
                for entry in feed.entries[:4]:
                    pub = None
                    if hasattr(entry, 'published_parsed') and entry.published_parsed:
                        pub = datetime(*entry.published_parsed[:6], tzinfo=timezone.utc)
                    source = entry.source.title if hasattr(entry, 'source') and hasattr(entry.source, 'title') else 'Unknown'
                    all_news.append({
                        'title': entry.get('title', 'N/A'),
                        'source': source,
                        'published': pub,
                        'link': entry.get('link', ''),
                        'category': category,
                    })
            except:
                pass
            time.sleep(0.3)

    # Deduplicate and filter last 48 hours
    seen = set()
    filtered = []
    cutoff = datetime.now(timezone.utc) - timedelta(hours=48)
    for item in all_news:
        title_key = item['title'].lower().strip()
        if title_key not in seen:
            seen.add(title_key)
            if item['published'] is None or item['published'] >= cutoff:
                filtered.append(item)

    # Sort by date
    filtered.sort(key=lambda x: x['published'] or datetime.min.replace(tzinfo=timezone.utc), reverse=True)

    # Build HTML
    if filtered:
        news_rows = ""
        current_cat = None
        for item in filtered:
            if item['category'] != current_cat:
                current_cat = item['category']
                news_rows += f"""
                <tr><td colspan='3' style='padding:12px 8px 4px;font-weight:bold;font-size:14px;
                    color:#4ecdc4;border-bottom:1px solid #4ecdc4;'>{current_cat}</td></tr>"""

            date_str = item['published'].strftime('%b %d, %H:%M UTC') if item['published'] else '—'
            news_rows += f"""
            <tr style='border-bottom:1px solid #2a2a3e;'>
                <td style='padding:6px 8px;color:#aaa;font-size:11px;white-space:nowrap;'>{date_str}</td>
                <td style='padding:6px 8px;'><a href="{item['link']}" style='color:#ddd;text-decoration:none;'>{item['title']}</a></td>
                <td style='padding:6px 8px;color:#888;font-size:11px;'>{item['source']}</td>
            </tr>"""

        display(HTML(f"""
        <div style='background:#1a1a2e;color:white;padding:20px;border-radius:10px;margin:10px 0;'>
            <h3 style='margin:0 0 5px 0;'>Live News Feed</h3>
            <p style='color:#aaa;font-size:11px;margin:0 0 15px 0;'>Last refreshed: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')} | Showing last 48 hours | {len(filtered)} headlines</p>
            <table style='color:white;width:100%;font-size:12px;border-collapse:collapse;'>
                {news_rows}
            </table>
        </div>
        """))
    else:
        print("No recent news found in the last 48 hours.")
else:
    display(Markdown("""**feedparser not installed.** Run `pip install feedparser` to enable the live news feed.

Manual news links:
- [Google News: TTF gas](https://news.google.com/search?q=TTF%20gas%20price%20Europe)
- [Google News: Ras Laffan](https://news.google.com/search?q=Ras%20Laffan%20LNG%20Qatar)
- [Google News: Strait of Hormuz](https://news.google.com/search?q=Strait%20of%20Hormuz%20shipping)
- [Google News: Cheniere Energy](https://news.google.com/search?q=Cheniere%20Energy%20LNG)
- [Google News: Brent crude](https://news.google.com/search?q=Brent%20crude%20oil%20Hormuz)
"""))
```
